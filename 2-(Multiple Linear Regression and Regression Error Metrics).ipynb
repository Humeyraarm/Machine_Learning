## Adjusted R2 score 

Means that how well is the relationship btw samples and features.Nearby R2 is better.Features more than need is reduce the R2 and so adjusted R2.

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]   # number of observations
    p = df.shape[1]-1 # number of independent variables 
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)
```

## What is the concept of punishment for RMSE error metric?

```python
variables = {"Actual": [1000, 2000, 3000, 4000, 5000], "pred": [1100, 2200, 3200, 4200, 5800]}
df_2 = pd.DataFrame(variables)
df_2["residual"] = abs(df_2.Actual - df_2.pred)  # abs = absolute value
df_2
```

```python
#mae
df_2.residual.sum()/5 
```

```python
#rmse
((df_2.residual**2).sum()/5)**0.5 
```

```python
variables2 = {"Actual": [1000, 2000, 3000, 4000, 5000], "pred": [1100, 2200, 3200, 4200, 7000]}
df_3 = pd.DataFrame(variables2)
df_3["residual"] = abs(df_3.Actual - df_3.pred)  # abs = absolute value
df_3
```

```python
#mae
df_3.residual.sum()/5 
```

```python
#rmse                               
((df_3.residual**2).sum()/5)**0.5 
```

**Bigger the residual higher difference btw mae and rmse.In RMSE, litte residuals more decrease,high residuals more increase,means punishment for RMSE eror metric.**

## Evaluation Model

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance:")
    print("--------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```
