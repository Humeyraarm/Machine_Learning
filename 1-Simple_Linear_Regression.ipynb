## Linear Regression

```python
sns.regplot(x="total_spend",y="sales",data=df,ci=None);  #ortalama karesel hatalarÄ± minimize ederek regression line Ã§izer.
```

## Splitting the dataset into X(independent variables) and y (dependent variable)

```python
# y_pred = b1X + b0
```

```python
X= df["total_spend"]
y= df["sales"]
```

## Determination of coefficients (slope and intercept)

```python
slope, intercept = np.polyfit(X, y, deg=1)
print("slope    :", slope)
print("intercept:", intercept)
```

## Why do we use the least squares error method to find the regression line that best fits the data?

```python
b1, b0 = np.polyfit(X, y, deg=1)
print("b1 :", b1)
print("b0 :", b0)
```

```python
y_pred = b1*X + b0
```

```python
values = {"actual": y, "predicted": y_pred, "residual":y-y_pred, "LSE": (y-y_pred)**2}   #residual = error regression doÄŸru Ã§alÄ±ÅŸÄ±yorsa,fit ediyorsa toplamlarÄ± sÄ±fÄ±r olur.
df_2 = pd.DataFrame(values)                                                              #Kare almamÄ±zÄ±n sebebi;gradientin tÃ¼rev almasÄ±nÄ± saÄŸlar.
df_2                                                                                     #bÃ¼yÃ¼k hata daha da bÃ¼yÃ¼r,kÃ¼Ã§Ã¼k hata daha da kÃ¼Ã§Ã¼lÃ¼r.
                                                                                         
```

```python
df_2.residual.sum().round()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(df_2["actual"],df_2["predicted"] )
print('r2 score for perfect model is', r2)
```
