## Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(["Outcome"], axis=1)
y=df["Outcome"]

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.20, 
                                                    stratify=y, 
                                                    random_state=42)
```

## Modelling and Model Performance

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
    y_train_pred = model.predict(X_train)
    y_pred = model.predict(X_test)
    
    print("Test_Set")
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    print()
    print("Train_Set")
    print(confusion_matrix(y_train, y_train_pred))
    print(classification_report(y_train, y_train_pred))
```

### Without Scalling

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
eval_metric(svm_model, X_train, y_train, X_test, y_test)

# SVM distance tabanlı bir model olduğundan datanın scale edilmesi gerekiyor.
# Ama sonuçların nasıl değiştiği görmek maksadıyla önce scale 
# etmeden devam edeceğiz.
```

```python
from sklearn.model_selection import cross_validate

model = SVC()

scores = cross_validate(model, 
                        X_train, 
                        y_train, 
                        scoring=['accuracy', 'precision','recall','f1'],
                        cv = 10,
                        return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]
```

### With Scalling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
operations = [("scaler", StandardScaler()), ("SVC", SVC())]
pipe_model = Pipeline(steps=operations)

# Şimdi scale yaparak scorları gözlemleyelim.
```

```python
pipe_model.fit(X_train, y_train)
eval_metric(pipe_model, X_train, y_train, X_test, y_test)

# Yukardaki skorlara bakarsanız ciddi farklar olduğunu göreceksiniz. 
# Scale yapmanın önemi daha iyi anlaşılıyor.
```

```python
operations = [("scaler", StandardScaler()), ("SVC", SVC())]
pipe_model = Pipeline(steps=operations)

scores = cross_validate(pipe_model, 
                        X_train, 
                        y_train, 
                        scoring=['accuracy', 'precision','recall','f1'],
                        cv = 10,
                        return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]

# Overfiting kontrolünü burada sağlıyoruz.
```

### With Best Parameters (GridsearchCV)

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'SVC__C': [0.001, 0.05, 0.01, 0.1],
              'SVC__gamma': ["scale", "auto", 0.2, 0.3],
              'SVC__kernel': ['rbf', 'linear'],
              'SVC__class_weight': ["balanced", None]}

# Hyper-parameter uzayımızı belirliyoruz.
```

```python
operations = [("scaler", StandardScaler()), ("SVC", SVC(probability=True))]
pipe_model = Pipeline(steps=operations)

svm_model_grid = GridSearchCV(pipe_model, 
                              param_grid, 
                              scoring="recall", 
                              cv=10, 
                              return_train_score=True)

# SVC predict_proba üzerinden prediction yapmaz. Tahminleri decision_function 
# üzerinden yapar. Eğer predict_proba almak istiyorsak SVC'nin probabilty 
# hyper_parametresini True olarak düzenlememiz gerekiyor.
# Decision_funksionın ne olduğuna aşağıda değineceğiz.
```

```python
svm_model_grid.fit(X_train, y_train)
```

```python
svm_model_grid.best_estimator_
```

```python
pd.DataFrame(svm_model_grid.cv_results_).loc[svm_model_grid.best_index_, ["mean_test_score", "mean_train_score"]]

#overfiting kontrolümüzü yapıyoruz.
```

```python
eval_metric(svm_model_grid, X_train, y_train, X_test, y_test)
```

```python
operations = [("scaler", StandardScaler()), 
              ("SVC", SVC(C= 0.05, gamma= 0.3, kernel= 'rbf', probability=True, 
                                                        class_weight="balanced"))]
model = Pipeline(steps=operations)

scores = cross_validate(model, 
                        X_train, 
                        y_train, 
                        scoring=['accuracy', 'precision','recall','f1'], 
                        cv = 10,
                        return_train_score=True)
df_scores = pd.DataFrame(scores, index = range(1, 11))
df_scores.mean()[2:]

# gridsearch CV sonrası extradan tekrar CV yapmaya gerek yok. 
# Ancak tüm metricleri görmek istiyorsanız tabi ki CV yapabilirsiniz
```
